```python
from sklearn.linear_model import LinearRegression
# from sklearn.tree import DecisionTreeRegressor
#if its decisionn tree.. linar unaichange to it,,,then im.... decision tree
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import numpy as np
from sklearn.externals import joblib

#splitting model into target and features
x=df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y= df['PRICE']

#x.head()

#splitting to y test and y train...folow hiyo order
x_train,x_test,y_train,y_test = train_test_split(x ,y, test_size=0.3, random_state=5)

#a variable created
# lr= DecisionTreeRegressor()
lr= LinearRegression()
lr.fit(x_train,y_train)
pr = lr.predict(x_test)
print(pr[0:4])

mae=metrics.mean_absolute_error(y_test,pr)
mse= metrics.mean_squared_error(y_test,pr)
rmse= np.sqrt(metrics.mean_absolute_error(y_test,pr))
print('MAE IS:',mae)
print('MSE IS:',mse)
print('RMSE IS:',rmse)
print(pr)
#save results to a csv file


result = pd.DataFrame(pr)
result.to_csv("My linear regression Output.csv")
# kupackage ni  unpack kwa flask
joblib.dump(lr,'My Linear Model.pkl')
print('THE RESULTS ARE:',result)

import matplotlib.pyplot as plt
plt.scatter(y_test,pr)
plt.show()
```
